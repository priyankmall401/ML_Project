```python
#dropping the unnecessary columns from dataframe
df.drop(['Unnamed: 0','id'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#separating numerical and categorical columns
num_df=df.select_dtypes(['int64','float64'])
num_df
```

```python
cat_df=df.select_dtypes(['object'])
cat_df
```

```python
#applying label encoding on cat_df
from sklearn.preprocessing import LabelEncoder
for i in cat_df:
    le=LabelEncoder()
    cat_df[i]=le.fit_transform(cat_df[i])
```

```python
cat_df
```

```python
#creating new dataframe by concatanating num_df and cat_df after label encoding on cat_df
df1=pd.concat([num_df,cat_df],axis=1)
df1.head()
```

```python
#checking for null values in new dataframe
df1.isnull().sum()
```

```python
#separating input features and output features
x=df1.iloc[:,0:22].values
x
```

```python
x.ndim
```

```python
y=df1.iloc[:,22:].values
y
```

```python
y.ndim
```

```python
#splitting the dataset into train and test data
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
#normalizing the data using minmax scaler
from sklearn.preprocessing import MinMaxScaler
```

```python
mn=MinMaxScaler()
```

```python
x_train=mn.fit_transform(xtrain)
y_train=mn.fit_transform(ytrain)
```

```python
x_test=mn.fit_transform(xtest)
y_test=mn.fit_transform(ytest)
```

```python
#applying logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
#model fitting
lr.fit(x_train,y_train)
```

```python
#predicting values
ypred1=lr.predict(x_train)
ypred1
```

```python
#finding model accuracy
from sklearn.metrics import accuracy_score
acc1=accuracy_score(y_train,ypred1)
print('logistic regression accuracy: ',acc1)
```

```python
# finding training accuracy
train_acc1=lr.score(x_train,y_train)
print('logistic regression training accuracy: ',train_acc1)
```

```python
#finding testing accuracy
test_acc1=lr.score(x_test,y_test)
print('logistic regression testing accuracy: ',test_acc1)
```

```python
#importing confusion matrix,classification report
from sklearn.metrics import confusion_matrix,classification_report
```

```python
#confusion matrix
con_matrix=confusion_matrix(y_train,ypred1)
print(con_matrix)
```

```python
#classification report
class_report=classification_report(y_train,ypred1)
print(class_report)
```
